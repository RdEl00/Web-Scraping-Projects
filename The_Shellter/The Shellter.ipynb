```python
# Importing necessary libraries for the project
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import The Shellter Website Dataset - Script available in the git repo.
df = pd.read_csv('turtle_data.csv')
```

```python
df.info()
```

### Data Wrangling and Tidying

```python
#removing the new line character (\n)
# Even rows follow the same pattern
rm = []
for x in range(0,12,2):
    rm.append(x)
print(rm)
```

```python
df.loc[[0, 2, 4, 6, 8, 10]]
```

```python
data = df.drop(rm)
```

```python
#Clean dataset, but still needs to manipulate it further to ease the analysis.
data
```

```python
#Delete duplicates & reset insex column
data.drop("Unnamed: 0",axis=1,inplace=True)
data.reset_index(drop=True)
```

```python
#Transpose the table
data1 = data.transpose()
```

```python
data1
```

```python
#Rename columns
data1.rename(columns={1:'Age', 3:'Weight', 5:'Sex', 7:'Breed', 9:'Source'},index=None, inplace=True)
```

```python
data1
```

```python
#clean table values
#Age column
data1['Age'] = data1['Age'].str.replace('AGE:', '')
data1['Age'] = data1['Age'].str.replace('Years Old', '')
data1['Age'] = data1['Age'].str.replace('Year Old', '')
#Weight column
data1['Weight'] = data1['Weight'].str.replace('WEIGHT:', '')
data1['Weight'] = data1['Weight'].str.replace('lbs', '')
data1['Weight'] = data1['Weight'].str.replace('lb', '')
#Sex Column
data1['Sex'] = data1['Sex'].str.replace('SEX:', '')
#Breed column
data1['Breed'] = data1['Breed'].str.replace('BREED:', '')
data1['Breed'] = data1['Breed'].str.replace('Tortoise', '')
data1['Breed'] = data1['Breed'].str.replace('Turtle', '')
#Source column
data1['Source'] = data1['Source'].str.replace('SOURCE:', '')
```

```python
#data1[["Age", "Weight"]] = data1[["Age", "Weight"]].apply(data1.to_numeric)
data1 = data1.astype({"Age": float})
data1 = data1.astype({"Weight": float})

data1.info()
```
